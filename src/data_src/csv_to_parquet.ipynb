```python
df_calendar = df_calendar.rename(columns={
    'wm_yr_wk': 'week_id',
    'weekday': 'weekday_name',
    'wday': 'weekday_number',
    'month': 'month_number',
    'year': 'year_number',
    'd': 'day_id',
    'event_name_1': 'event_name_primary',
    'event_type_1': 'event_type_primary',
    'event_name_2': 'event_name_secondary',
    'event_type_2': 'event_type_secondary',
    'snap_CA': 'snap_california',
    'snap_TX': 'snap_texas',
    'snap_WI': 'snap_wisconsin'
})

df_calendar['date'] = pd.to_datetime(df_calendar['date'])
```

### Monster

```python
import pandas as pd
from pathlib import Path

Path("monster_parts").mkdir(exist_ok=True)

id_vars = [
    "series_id","product_id","department_id",
    "category_id","store_id","state_code"
]

chunks = []
for store, df_store in df_sales.groupby("store_id"):
    print(f"Processing store {store} …")
    df_long = df_store.melt(id_vars=id_vars, var_name="day_id", value_name="sales")
    df_long = df_long.merge(df_calendar, how="left", on="day_id")
    df_long = df_long.merge(df_sell_prices, how="left",
                            on=["store_id","product_id","week_id"])
    df_long = df_long.sort_values(["product_id", "day_id"])
    fn = Path("monster_parts") / f"monster_{store}.parquet"
    df_long.to_parquet(fn, compression="snappy", index=False)
    print("   ✔ saved", fn)
```

```python
from pathlib import Path

folder = Path("monster_parts")

for f in sorted(folder.glob("monster_*.parquet")):
    size_mb = f.stat().st_size / 1024 ** 2
    print(f"{f.name:<35}  {size_mb:8.1f} MB")
```

```python
import pandas as pd

for f in sorted(folder.glob("monster_*.parquet")):
    n = len(pd.read_parquet(f, columns=["sales"]))   # read just one column
    print(f"{f.name:<35}  {n:>10,} rows")
```

```python
files = sorted(folder.glob("monster_*.parquet"))
total_rows = sum(
    len(pd.read_parquet(f, columns=["sales"])) for f in files
)
print(f"Total rows across {len(files)} parts: {total_rows:,}")
```
